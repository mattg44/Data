### Linear regression method

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
df_reg = df.drop("is_genuine", axis=1)
df_reg.head()
```

```python
testdf = df_reg[df_reg['margin_low'].isnull()==True]
traindf = df_reg[df_reg['margin_low'].isnull()==False]
X = traindf.drop('margin_low',axis=1)
y = traindf['margin_low']
```

```python
lr.fit(X, y)
testdf.drop("margin_low", axis=1, inplace=True)
testdf['margin_low'] = lr.predict(testdf)
```

```python
df_final = pd.concat([traindf, testdf], axis=0)
```

```python
df_final.info()
```

```python
traindf.shape
```

```python
y.shape
```

```python
#calcul du R²
lr.score(X, y)
```

```python
RMSE=np.sqrt(((y-lr.predict(X))**2).sum()/len(y))
RMSE
```

```python
# deux graphiques (qu'il faudrait mieux préparer) représentant : les valeurs de y en fonction des
# valeurs prédites avec le modèle de régresssion linéaire et les valeurs de Y en fonction des résidus.
plt.plot(y, lr.predict(traindf.drop("margin_low", axis=1)), ".")
plt.show()

plt.plot(y, y - lr.predict(traindf.drop("margin_low", axis=1)), ".")
plt.show()
```

# Prediction

```python
from sklearn import metrics
from sklearn.metrics import accuracy_score
```

```python
X = df_final2.drop(["is_genuine"], axis=1)
y = df_final2["is_genuine"]
```

```python
# scaler = MinMaxScaler()
# X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
import lazypredict
from lazypredict.Supervised import LazyClassifier
```

```python
clf = LazyClassifier(verbose=0, ignore_warnings=True)
pd.options.display.max_columns = 30
models, predictions = clf.fit(X_train, X_test, y_train, y_test)
print(models)
```

```python
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
```

```python
print("Evaluation metrics - Random forest:")
print("Accuracy: " + str(metrics.accuracy_score(y_test, y_pred)))
print("F1 score: " + str(metrics.f1_score(y_test, y_pred, average="macro")))
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)
```

```python
print("Evaluation metrics - Logistic regression:")
print("Accuracy: " + str(metrics.accuracy_score(y_test, y_pred)))
print("F1 score: " + str(metrics.f1_score(y_test, y_pred_lr, average="macro")))
```

```python
from sklearn.metrics import confusion_matrix
 
#Generate the confusion matrix
rf_matrix = confusion_matrix(y_test, y_pred)

print(rf_matrix)
```

```python
plt.figure(figsize=(8, 8))
ax = sns.heatmap(rf_matrix, annot=True, fmt='.3g')
ax.set_title('Confusion Matrix with labels\n\n')
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ')
ax.xaxis.set_ticklabels(['False','True'])
ax.yaxis.set_ticklabels(['False','True'])
plt.show();
```

```python
#Generate the confusion matrix
lr_matrix = confusion_matrix(y_test, y_pred_lr)

print(lr_matrix)
```

```python
plt.figure(figsize=(8, 8))
sns.heatmap(lr_matrix, annot=True, fmt='.3g')
```
